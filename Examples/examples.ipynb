```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = 255*np.ones((1200 , 1200 , 3 ) , dtype = np.uint8);
row , cloum , _ = img.shape
j = 0
thickness = 4
r , g , b = (255 , 0 , 0) , (0 , 255 , 0) , (0 , 0 , 255)

for i in range (0 , cloum - 1 , 99):
    
    pt1 =(i , cloum)
    pt2 =(0 , i)
    cv2.line(img , pt1 , pt2 , r , thickness)
    
    cv2.line(img , np.flip(pt1) , np.flip(pt2) , b , thickness)
    
    pt1 = (i, cloum - 1)
    pt2 = (row - 1, cloum - i)
    cv2.line(img, pt1, pt2, g, thickness)
    
    pt1 = (row - i, 0)
    pt2 = (0, i)
    cv2.line(img, pt1, pt2, g, thickness)

    j+=1
    
cv2.destroyAllWindows()
plt.imshow(img)
plt.axis("off");
```

```python
img = 255*np.ones((600 , 600 , 3 ) , dtype = np.uint8);
row , cloum , _ = img.shape
j = 0
thickness = 3
r , g , b = (255 , 0 , 0) , (0 , 255 , 0) , (0 , 0 , 255)

for i in range (0 , int(cloum/2), 50):
    
    pt1 =(i , 300)
    pt2 =(300 , 300 + i )
    cv2.line(img , pt1 , pt2 , g , thickness)
    cv2.line(img , np.flip(pt1) , np.flip(pt2) , g , thickness)
    
    pt1 =(300 , i)
    pt2 =(300 - i , 300 )
    cv2.line(img , pt1 , pt2 , r , thickness)
    
    pt1 = (300, 300+i)
    pt2 = (600-i, 300)
    cv2.line(img, pt1, pt2, b, thickness)    
    cv2.line(img , np.flip(pt1) , np.flip(pt2) , b , thickness)

    j+=1
    
cv2.destroyAllWindows()
plt.imshow(img)
plt.axis("off");
```

```python
img = 255*np.ones((1200 , 1200) ,  dtype = np.uint8);
black = (0 , 0 , 0)
j=0

for i in range (1 , 24):
    center = (600 , 600)
    thickness =24 - i
    cv2.circle(img , center , i*26 , black , thickness)
    
    j+=1

cv2.destroyAllWindows()
plt.imshow(img , cmap="gray" , vmin=0 , vmax=255)
plt.axis("off");
```
